# XGBoost 텍스트 유사도 분석 모델

```python
import numpy as np
import pandas as pd
import os
import json

from sklearn.model_selection import train_test_split

DATA_IN_PATH = r'/data_in/'

train_q1_data = np.load(open(DATA_IN_PATH + 'q1_train.npy', 'rb'))
train_q2_data = np.load(open(DATA_IN_PATH + 'q2_train.npy', 'rb'))
train_labels = np.load(open(DATA_IN_PATH + 'label_train.npy', 'rb'))
```

```python
train_input = np.stack((train_q1_data, train_q2_data), axis=1)
print(train_input.shape)
```

```python
train_input, eval_input, train_label, eval_label = train_test_split(train_input, train_labels,
                                                                    test_size=0.2, random_state=4242)
```

모델 구성

```python
import xgboost as xgb

train_data = xgb.DMatrix(train_input.sum(axis=1), label=train_label)
eval_data = xgb.DMatrix(eval_input.sum(axis=1), label=eval_label)

data_list = [(train_data, 'train'), (eval_data, 'valid')]
```

```python
params = {} # 인자를 통해 xgb모델에 넣는다
params['objective'] = 'binary:logistic'
params['eval_metric'] = 'rmse' #root mean squared error를 평가 지표로 사용

bst = xgb.train(params, train_data, num_boost_round=1000, evals=data_list, early_stopping_rounds=10)
```

예측하기

```python
test_q1_data = np.load(open(DATA_IN_PATH + 'test_q1.npy', 'rb'))
test_q2_data = np.load(open(DATA_IN_PATH + 'test_q2.npy', 'rb'))
test_id_data = np.load(open(DATA_IN_PATH + 'test_id.npy', 'rb'))
```

```python
test_input = np.stack((test_q1_data, test_q2_data), axis=1)
test_data = xgb.DMatrix(test_input.sum(axis=1))
test_predict = bst.predict(test_data)
```
